```python
gsod_mergeddf = pd.read_csv("LatestData/merged-file-2010.txt",delim_whitespace=True,
               header=None,
               names=[
                   "station",
                   "wban",
                   "yearday",
                   "temp",
                   "temp2",
                   "dewp",
                   "dewp2",
                   "slp",
                   "slp2",
                   "stp",
                   "stp2",
                   "visib",
                   "visib2",
                   "wdsp",
                   "wdsp2",
                   "mxspd",
                   "gust",
                   "max",
                   "min",
                   "prcp",
                   "sndp",
                   "frshtt"
               ])
```

```python
gsod_mergeddf = gsod_mergeddf[gsod_mergeddf['station'] != '999999']
gsod_mergeddf = gsod_mergeddf.set_index('station')
gsod_mergeddf.head()
```

```python
mergedgsodcounty = gsod_mergeddf.merge(gsodcountydf, left_index=True, right_index=True)
mergedgsodcounty.head()
```

```python
mergedgsodcounty.to_csv('LatestData/gsod_2010.csv')
```
